We understand the following facts:
- All features and the target y are quite correlated with the market. To maximise performance, we will train 4 separate regressors for each market.
- The x3A...x3E are heavily correlated. We will decorrelate them by using a PCA. 
- No "seasonal" dependence is observed (y is stable with day). So, time is ignored

## Feature decorrelation (PCA)

```python
#Convert to numpy arrays
features = list(df_train)
for f in ['y','Market','Day','Stock','Weight']:
    features.remove(f)
    
x_raw_train = df_train[features]
x_raw_eval = df_eval[features]
```

```python
#Run PCA (only for x3A...x3E)
pca = PCA()
x_new_train = pca.fit_transform(x_raw_train)
x_new_eval = pca.fit_transform(x_raw_eval)
```

```python
#Update dataframes
for i, f in enumerate(features):
    df_train[f] = x_new_train[:,i]
    df_eval[f] = x_new_eval[:,i]
```

## Build XGBOOST regressor (based on gradient-boosted DT's)

```python
#Recast data properly

features = list(df_train)
for f in ['y','Day','Weight']:
    features.remove(f)

#target
y_train_full = df_train["y"]
    
#features
X_train_full = df_train[features]
    
#sample weight
w_train_full = df_train["Weight"]
    
#split
X_train, X_test, y_train, y_test, w_train, w_test = train_test_split(X_train_full,
                                                                     y_train_full,
                                                                     w_train_full,
                                                                     test_size=0.33, random_state=42)
```

```python
#Run cross-validation
parameters = {'max_depth': [3,5,7],
              'n_estimators': [300,1000,2000]}
    
xgb = XGBRegressor(random_state=42)

xgb_grid = GridSearchCV(xgb,
                        parameters,
                        cv = 3,
                        n_jobs = 4,
                        fit_params = {'sample_weight':w_train}, 
                        verbose=True)

xgb_grid.fit(X_train, y_train)
```

```python
print "Best parameters"
print xgb_grid.best_params_
```

```python
#Run nominal training using best config
kwargs = xgb_grid.best_params_
kwargs['sample_weight'] = w_train
xgb = XGBRegressor(random_state=42, **kwargs)
xgb.fit(X_train, y_train)
xgb.get_booster().save_model("../weights/xgboost_withWeights_PCA.xgb")
```

```python
#Compare training and test wMSE
y_p_train = xgb.predict(X_train)
y_p_test = xgb.predict(X_test)
wMSE_train = wMSE(w_train, y_train, y_p_train)
wMSE_test = wMSE(w_test, y_test, y_p_test)
print "Training wMSE"
print wMSE_train
print "Test wMSE"
print wMSE_test
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111)
plot_importance(xgb, ax=ax);
plt.show()
```

```python
#Predict y on evaluation set
y_p_eval = pd.Series(data=np.array([]), dtype='float64')

features = list(df_eval)
features.remove('Day')
X_eval = df_eval[features]
y_p_eval = pd.Series(data=xgb.predict( X_eval ), dtype='float64')
```

```python
y_p_eval = y_p_eval.rename('y')
y_p_eval.index.name = 'Index'
y_p_eval.head()
```
